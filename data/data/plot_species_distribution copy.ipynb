```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
data_file = "all_DIB_image_data.csv"
df = pd.read_csv(data_file, header=0)
df.head()
```

```python
# Basic stats
total_species = df.shape[0]
total_images = df['nb_images'].sum()
expected_species = 10922
expected_images = 4876536
print("total species: ", total_species, "--- total images: ", total_images)
print(f"Expected nb_species: {expected_species} --- expected nb_images: {expected_images}")
print("Missing species: ", expected_species - total_species, " --- missing images: ", expected_images - total_images)
```

### Nb species per parts

```python
images_per_part_df = df.groupby('dataset_part').sum().reset_index() 
sns.barplot(data=images_per_part_df, y='nb_images', x='dataset_part')
plt.xticks(rotation=90)
plt.title("Nb images per dataset part")
```

### Species per part

```python
species_per_part_df = df.groupby('dataset_part').count()
species_per_part_df.head()
```

```python
species_per_part_df = df.groupby('dataset_part').count().reset_index()
sns.barplot(data=species_per_part_df, y='bird_name', x='dataset_part')
plt.xticks(rotation=90)
plt.ylabel("nb_species")
plt.title("Nb species per dataset part")
```

### Images per species

```python
sns.histplot(data=df, x="nb_images")
plt.title("Nb image per species")
```
