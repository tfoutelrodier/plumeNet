```python
import pandas as pd
import json
```

```python
df = pd.read_csv('french_birds_metadata.csv', sep=';')
df.head()
```

```python
# Sélectionner les colonnes nom_LT, name et URL
df_usefull = df[['Nom_LT', 'Url_FR', 'Nom_FR']]

# Convertir le DataFrame en un dictionnaire Python
data = df_usefull.to_dict(orient='records')

# Créer un nouveau dictionnaire Python vide
json_data = {}
 
# Parcourir le dictionnaire Python et ajouter chaque enregistrement au nouveau dictionnaire
for item in data:
    json_data[item['Nom_LT'].lower().replace(' ', '_')] = {'url': item['Url_FR'], 'Nom_LT': item['Nom_LT'], 'Nom_FR': item['Nom_FR']}

# Convertir le nouveau dictionnaire Python en un fichier JSON
with open('names_and_urls.json', 'w') as f:
    json.dump(json_data, f)
```
